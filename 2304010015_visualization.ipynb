```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('historical_automobile_sales.csv')
df.head(n=10)
```

```python
df_yearly = df.groupby('Year')['Automobile_Sales'].mean()
```

```python
tahun_resesi = df[df['Recession'] == 1]['Year'].unique()
print(tahun_resesi)
```

```python
plt.figure(figsize=(15,8))
ax = df_yearly.plot(kind='line', marker='o', color='blue')

ax.xaxis.set_ticks(df_yearly.index)
ax.tick_params(axis='x', rotation=45)

plt.annotate('Recession 1980',
             xy=(1980, df_yearly[1980]),
             xytext=(1980, df_yearly[1980] + 60),
             arrowprops=dict(facecolor='red', shrink=0.05, width=1.5, headwidth=6),
             color='red', weight='bold', ha='center')
plt.annotate('Recession 2000',
             xy=(2000, df_yearly[2000]),
             xytext=(2000, df_yearly[2000] + 60),
             arrowprops=dict(facecolor='red', shrink=0.05, width=1.5, headwidth=6),
             color='red', weight='bold', ha='center')
plt.annotate('Recession 2020',
             xy=(2020, df_yearly[2020]),
             xytext=(2020, df_yearly[2020] + 60),
             arrowprops=dict(facecolor='red', shrink=0.05, width=1.5, headwidth=6),
             color='red', weight='bold', ha='center')

plt.title('Automobile Sales Fluctuation from Year to Year', fontsize=14, weight='bold', pad=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Automobile Sales', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show
```
